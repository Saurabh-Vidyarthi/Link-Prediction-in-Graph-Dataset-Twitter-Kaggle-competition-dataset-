```python
G_complete_train_data = nx.from_pandas_edgelist(df, source='Form', target='To', create_using=nx.DiGraph())
```

```python
# Calculate indegree and outdegree for each node
indegree = dict(G_complete_train_data.in_degree())
outdegree = dict(G_complete_train_data.out_degree())
```

```python
print(len(indegree))
len(outdegree)
```

```python
src_indegree_df = pd.DataFrame(list(indegree.items()), columns=['Node', 'src_Indegree'])
src_outdegree_df = pd.DataFrame(list(outdegree.items()), columns=['Node', 'src_Outdegree'])
```
