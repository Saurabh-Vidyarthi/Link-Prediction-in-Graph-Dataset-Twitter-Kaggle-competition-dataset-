```python
import warnings
warnings.filterwarnings("ignore")

import csv
import pandas as pd#pandas to create small dataframes 
import datetime #Convert to unix time
import time #Convert to unix time
# if numpy is not installed already : pip3 install numpy
import numpy as np#Do aritmetic operations on arrays
# matplotlib: used to plot graphs
import matplotlib
import matplotlib.pylab as plt
import seaborn as sns#Plots
from matplotlib import rcParams#Size of plots  
from sklearn.cluster import MiniBatchKMeans, KMeans#Clustering
import math
```

```python
train_data = pd.read_csv('train_data_with_jacard_corrected.csv')
train_data
```

```python
X = train_data[['Form','To']]
X
type(X)
```

```python
whole_data = pd.read_csv('data_int.csv')
whole_data = whole_data.rename(columns={'0': 'Form', '1': 'To'})
whole_data
```

```python
G = nx.DiGraph()

for _, row in whole_data.iterrows():
    G.add_edge(row['Form'], row['To'])
```

```python
# Function to calculate shortest path length
def calculate_shortest_path_length(row):
    source_node = row['Form']
    target_node = row['To']
    
    try:
        shortest_path_length = nx.shortest_path_length(G, source=source_node, target=target_node)
        return shortest_path_length
    except nx.NetworkXNoPath:
        return float(10000.0)  # Return infinity if no path exists

# Apply the function to each row in the DataFrame and append the results as a new column
X['shortest_path_length'] = X.apply(calculate_shortest_path_length, axis=1)
```

```python
Y = X['shortest_path_length']
Y
```

```python
train_data
```

```python
train_data['shortest_path'] = Y
train_data
```

```python
file_path = 'train_data_with_sh_path_n_all_ft.csv'

# Save the DataFrame to a CSV file
train_data.to_csv(file_path, index=False)
```

```python
test_data = pd.read_csv('kaggle_test_data_with_jacard_crrt.csv')
test_data
```

```python
X_1 = test_data[['Form','To']]
X_1
```

```python
def calculate_shortest_path_length(row):
    source_node = row['Form']
    target_node = row['To']
    
    if source_node not in G or target_node not in G:
        return 10000.0  # Return -1 if source or target node is not in the graph
    
    try:
        shortest_path_length = nx.shortest_path_length(G, source=source_node, target=target_node)
        return shortest_path_length
    except nx.NetworkXNoPath:
        return 10000.0  # Return a large value if no path exists

# Apply the function to each row in the DataFrame and append the results as a new column
X_1['shortest_path_length'] = X_1.apply(calculate_shortest_path_length, axis=1)
```
