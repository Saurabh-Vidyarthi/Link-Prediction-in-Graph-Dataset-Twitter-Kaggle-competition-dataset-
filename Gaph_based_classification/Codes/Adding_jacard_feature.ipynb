# Adding Jaccard feature to test data and train data

```python
#Importing Libraries
# please do go through this python notebook: 
import warnings
warnings.filterwarnings("ignore")

import csv
import pandas as pd#pandas to create small dataframes 
import datetime #Convert to unix time
import time #Convert to unix time
# if numpy is not installed already : pip3 install numpy
import numpy as np#Do aritmetic operations on arrays
# matplotlib: used to plot graphs
import matplotlib
import matplotlib.pylab as plt
import seaborn as sns#Plots
from matplotlib import rcParams#Size of plots  
from sklearn.cluster import MiniBatchKMeans, KMeans#Clustering
import math
import warnings
import networkx as nx
```

```python
train_data = pd.read_csv('concat_df_with_2_feature.csv')
train_data
```

```python
df = pd.read_csv('data_int.csv')
df
```

```python
df = df.rename(columns={'0': 'Form', '1': 'To'})
df
```

```python
train_graph=nx.read_edgelist('data_int.csv',delimiter=',',create_using=nx.DiGraph(),nodetype=int)
print(nx.info(train_graph))
```

```python
G_d1 = nx.Graph()

for _, row in df.iterrows():
    G_d1.add_edge(row['Form'], row['To'])


jaccard_coefficients = []

for _, row_d2 in train_data.iterrows():
    node1_d2 = row_d2['Form']
    node2_d2 = row_d2['To']
   
    jaccard = nx.jaccard_coefficient(G_d1, [(node1_d2, node2_d2)])
   
    for u, v, j in jaccard:
        jaccard_coefficients.append(j)

train_data['Jaccard_Coefficient'] = jaccard_coefficients
```

```python
def follows_back(a,b):
    if train_graph.has_edge(b,a):
        return 1
    else:
        return 0
```

```python
train_data['follow_back'] = train_data.apply(lambda row: follows_back(row['Form'],row['To']),axis=1)
train_data
```

```python
train_data.head(50)
```

```python
#weight for source and destination of each link
Weight_in = {}
Weight_out = {}
for i in  tqdm(train_graph.nodes()):
    s1=set(train_graph.predecessors(i))
    w_in = 1.0/(np.sqrt(1+len(s1)))
    Weight_in[i]=w_in
    
    s2=set(train_graph.successors(i))
    w_out = 1.0/(np.sqrt(1+len(s2)))
    Weight_out[i]=w_out
    
#for imputing with mean
mean_weight_in = np.mean(list(Weight_in.values()))
mean_weight_out = np.mean(list(Weight_out.values()))
```

```python
train_data['weight_in'] = train_data.To.apply(lambda x: Weight_in.get(x,mean_weight_in))
train_data['weight_out'] = train_data.Form.apply(lambda x: Weight_out.get(x,mean_weight_out))
train_data['weight_f1'] = train_data.weight_in + train_data.weight_out
train_data['weight_f2'] = train_data.weight_in * train_data.weight_out
train_data['weight_f3'] = (2*train_data.weight_in + 1*train_data.weight_out)
train_data['weight_f4'] = (1*train_data.weight_in + 2*train_data.weight_out)
train_data
```

```python
file_path = 'train_data_with_jacard_corrected.csv'

# Save the DataFrame to a CSV file
train_data.to_csv(file_path, index=False)
```
